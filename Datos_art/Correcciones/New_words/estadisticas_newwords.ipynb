```python
df_ab_a = pd.read_csv(e_ab_a + '.csv', index_col=0, dtype=str)
df_ac_a = pd.read_csv(e_ac_a + '.csv', index_col=0, dtype=str)
df_ad_a = pd.read_csv(e_ad_a + '.csv', index_col=0, dtype=str)
df_ae_a = pd.read_csv(e_ae_a + '.csv', index_col=0, dtype=str)

df_ab_b = pd.read_csv(e_ab_b + '.csv', index_col=0, dtype=str)
df_ac_b = pd.read_csv(e_ac_b + '.csv', index_col=0, dtype=str)
df_ad_b = pd.read_csv(e_ad_b + '.csv', index_col=0, dtype=str)
df_ae_b = pd.read_csv(e_ae_b + '.csv', index_col=0, dtype=str)

df_ab_c = pd.read_csv(e_ab_c + '.csv', index_col=0, dtype=str)
df_ac_c = pd.read_csv(e_ac_c + '.csv', index_col=0, dtype=str)
df_ad_c = pd.read_csv(e_ad_c + '.csv', index_col=0, dtype=str)
df_ae_c = pd.read_csv(e_ae_c + '.csv', index_col=0, dtype=str)

df_ab_d = pd.read_csv(e_ab_d + '.csv', index_col=0, dtype=str)
df_ac_d = pd.read_csv(e_ac_d + '.csv', index_col=0, dtype=str)
df_ad_d = pd.read_csv(e_ad_d + '.csv', index_col=0, dtype=str)
df_ae_d = pd.read_csv(e_ae_d + '.csv', index_col=0, dtype=str)

df_ab_e = pd.read_csv(e_ab_e + '.csv', index_col=0, dtype=str)
df_ac_e = pd.read_csv(e_ac_e + '.csv', index_col=0, dtype=str)
df_ad_e = pd.read_csv(e_ad_e + '.csv', index_col=0, dtype=str)
df_ae_e = pd.read_csv(e_ae_e + '.csv', index_col=0, dtype=str)


Conjunto_EN_A = [df_ab_a, df_ac_a, df_ad_a, df_ae_a]
Conjunto_FR_A = [df_ab_b, df_ac_b, df_ad_b, df_ae_b]
Conjunto_GE_A = [df_ab_c, df_ac_c, df_ad_c, df_ae_c]
Conjunto_IT_A = [df_ab_d, df_ac_d, df_ad_d, df_ae_d]
Conjunto_SP_A = [df_ab_e, df_ac_e, df_ad_e, df_ae_e]

Conjunto_EN_B = [df_ab_b, df_ab_c, df_ab_d, df_ab_e]
Conjunto_FR_B = [df_ab_a, df_ac_c, df_ac_d, df_ac_e]
Conjunto_GE_B = [df_ac_a, df_ac_b, df_ad_d, df_ad_e]
Conjunto_IT_B = [df_ad_a, df_ad_b, df_ad_c, df_ae_e]
Conjunto_SP_B = [df_ae_a, df_ae_b, df_ae_c, df_ae_d]

Etiqueta_EN_A = [e_ab_a, e_ac_a, e_ad_a, e_ae_a]
Etiqueta_FR_A = [e_ab_b, e_ac_b, e_ad_b, e_ae_b]
Etiqueta_GE_A = [e_ab_c, e_ac_c, e_ad_c, e_ae_c]
Etiqueta_IT_A = [e_ab_d, e_ac_d, e_ad_d, e_ae_d]
Etiqueta_SP_A = [e_ab_e, e_ac_e, e_ad_e, e_ae_e]

Etiqueta_EN_B = [e_ab_b, e_ab_c, e_ab_d, e_ab_e]
Etiqueta_FR_B = [e_ab_a, e_ac_c, e_ac_d, e_ac_e]
Etiqueta_GE_B = [e_ac_a, e_ac_b, e_ad_d, e_ad_e]
Etiqueta_IT_B = [e_ad_a, e_ad_b, e_ad_c, e_ae_e]
Etiqueta_SP_B = [e_ae_a, e_ae_b, e_ae_c, e_ae_d]
```

```python
def Palabras_unicas(conjuntos):
    decadas  = arange(1900, 2010, 10)
    palabras = []
    for conjunto in conjuntos:
        for decada in decadas:
            pdecada = list(conjunto[str(decada)].values)
            palabras += pdecada
    palabras_unicas = list(unique(array(palabras)))
    palabras_unicas.remove('-')
    return palabras_unicas

def Estadisticas_nuevas_A(unicas, conjuntos, etiquetas):
    Idioma_A  = []
    Idioma_B  = []
    Global_migradas_por_A  = []
    Total_recibidas_A_en_B = []
    Porcentaje_A_en_B      = [] 
    Decada_mayor_migracion = []
    Valor_decada_mayor_migracion      = []
    Porcentaje_decada_mayor_migracion = []
    decadas  = arange(1900, 2010, 10)
    for i in range(len(conjuntos)):
        conjunto = conjuntos[i]
        etiqueta = etiquetas[i]
        idiomaA  = etiqueta[:2]
        idiomaB  = etiqueta[-2:]
        palabras = []
        total_decada  = 0
        nombre_decada = 0
        for decada in decadas:
            subconjunto = set(list(conjunto[str(decada)]))
            pdecada  = list(subconjunto)
            palabras += pdecada
            tdecada  = len(pdecada)-1
            if tdecada > total_decada:
                total_decada  = tdecada
                nombre_decada = decada
        palabras_unicas = list(unique(array(palabras)))
        palabras_unicas.remove('-')
        Idioma_A.append(idiomaA)
        Idioma_B.append(idiomaB)
        Global_migradas_por_A.append(len(unicas))
        Total_recibidas_A_en_B.append(len(palabras_unicas))
        Porcentaje_A_en_B.append(round( (len(palabras_unicas)/len(unicas)) *100, 3 ) )
        Decada_mayor_migracion.append(nombre_decada)
        Valor_decada_mayor_migracion.append(total_decada)
        Porcentaje_decada_mayor_migracion.append(round( (total_decada/len(palabras_unicas)) *100, 3 ) )  
    df = pd.DataFrame()
    df['Idioma_A'] = Idioma_A
    df['Idioma_B'] = Idioma_B
    df['Global_migradas_por_A']   = Global_migradas_por_A
    df['Total_recibidas_A_en_B']  = Total_recibidas_A_en_B
    df['Porcentaje_A_en_B']       = Porcentaje_A_en_B
    df['Decada_mayor_migracion']  = Decada_mayor_migracion
    df['Valor_decada_mayor_migracion']      = Valor_decada_mayor_migracion
    df['Porcentaje_decada_mayor_migracion'] = Porcentaje_decada_mayor_migracion
    return df


def Estadisticas_nuevas_B(unicas, conjuntos, etiquetas):
    Idioma_A  = []
    Idioma_B  = []
    Global_recibidas_por_B = []
    Total_migradas_A_en_B  = []
    Porcentaje_A_en_B      = [] 
    Decada_mayor_recepcion = []
    Valor_decada_mayor_recepcion      = []
    Porcentaje_decada_mayor_recepcion = []
    decadas  = arange(1900, 2010, 10)
    for i in range(len(conjuntos)):
        conjunto = conjuntos[i]
        etiqueta = etiquetas[i]
        idiomaA  = etiqueta[:2]
        idiomaB  = etiqueta[-2:]
        palabras = []
        total_decada  = 0
        nombre_decada = 0
        for decada in decadas:
            subconjunto = set(list(conjunto[str(decada)]))
            pdecada  = list(subconjunto)
            palabras += pdecada
            tdecada  = len(pdecada)-1
            if tdecada > total_decada:
                total_decada  = tdecada
                nombre_decada = decada
        palabras_unicas = list(unique(array(palabras)))
        palabras_unicas.remove('-')
        Idioma_A.append(idiomaA)
        Idioma_B.append(idiomaB)
        Global_recibidas_por_B.append(len(unicas))
        Total_migradas_A_en_B.append(len(palabras_unicas))
        Porcentaje_A_en_B.append(round( (len(palabras_unicas)/len(unicas)) *100, 3 ) )
        Decada_mayor_recepcion.append(nombre_decada)
        Valor_decada_mayor_recepcion.append(total_decada)
        Porcentaje_decada_mayor_recepcion.append(round( (total_decada/len(palabras_unicas)) *100, 3 ) )  
    df = pd.DataFrame()
    df['Idioma_A'] = Idioma_A
    df['Idioma_B'] = Idioma_B
    df['Global_recibidas_por_B']  = Global_recibidas_por_B
    df['Total_migradas_A_en_B']   = Total_migradas_A_en_B
    df['Porcentaje_A_en_B']       = Porcentaje_A_en_B
    df['Decada_mayor_recepcion']  = Decada_mayor_recepcion
    df['Valor_decada_mayor_recepcion']      = Valor_decada_mayor_recepcion
    df['Porcentaje_decada_mayor_recepcion'] = Porcentaje_decada_mayor_recepcion
    return df
```

```python
Unicas_EN_A = Palabras_unicas(Conjunto_EN_A)
Unicas_FR_A = Palabras_unicas(Conjunto_FR_A)
Unicas_GE_A = Palabras_unicas(Conjunto_GE_A)
Unicas_IT_A = Palabras_unicas(Conjunto_IT_A)
Unicas_SP_A = Palabras_unicas(Conjunto_SP_A)

Unicas_EN_B = Palabras_unicas(Conjunto_EN_B)
Unicas_FR_B = Palabras_unicas(Conjunto_FR_B)
Unicas_GE_B = Palabras_unicas(Conjunto_GE_B)
Unicas_IT_B = Palabras_unicas(Conjunto_IT_B)
Unicas_SP_B = Palabras_unicas(Conjunto_SP_B)
```

```python
Cj_origen = [Conjunto_EN_A, Conjunto_FR_A, Conjunto_GE_A, Conjunto_IT_A, Conjunto_SP_A]
Et_origen = [Etiqueta_EN_A, Etiqueta_FR_A, Etiqueta_GE_A, Etiqueta_IT_A, Etiqueta_SP_A]
Un_origen = [Unicas_EN_A,   Unicas_FR_A,   Unicas_GE_A,   Unicas_IT_A,   Unicas_SP_A]
DF_A = pd.DataFrame()

for i in range(len(Un_origen)):
    lt_unicas    = Un_origen[i]
    lt_conjunto  = Cj_origen[i]
    lt_etiquetas = Et_origen[i]
    dn   = Estadisticas_nuevas_A(lt_unicas, lt_conjunto, lt_etiquetas)
    DF_A = pd.concat([DF_A, dn], axis=0, ignore_index=True)
```
