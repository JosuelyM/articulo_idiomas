# Calculo del retorno

El retorno es la variacion porcentual de un valor P en dos periodos de tiempo.

$$r_{t} = \frac{P_{i+1} - P_{i}}{P_{i}} $$

El retorno logaritmico es tambien usado junto al retorno "normal"
$$r_{t} = \ln\left(\frac{P_{i+1}}{P_{i}}\right)$$

Donde una de sus ventajas es el ser aditivo, además por su interpretación geometrica, $e^{r}$ es el factor de creicimento entre los dos valores.

Definido el retorno, hemos calculado tanto el  "normal" como el logaritmico los valores del Uso entre idiomas para dos temporalidades, cada 5 años y cada 10 años.

Por ejemplo para t=5, se ha calculado el retorno entre los valores del Uso en los años de 1900 y 1905, 1905 y 1910, 1910 y 1915 ........, 2000 y 2005.

Para t=10, se ha calculado el retorno entre los valores de Uso en los años de 1900 y 1910, 1910 y 1920, ......., 1990 y 2000.
